10.  Load the `iris` dataset that we used in class.  Use `iris` as the name of your data frame.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

11.  What is the correlation between the variables `sepal_length` and `sepal_width` for all the irises?

```python
iris[['sepal_length','sepal_width']].corr()

#.corr() - correlation - combined with dataframe iris calls the correlation between the sepal length and sepal width for all irises in the dataset
```

12.  Which of the three species has the highest correlation between `sepal_length` and `sepal_width`?  Show any calculations you use to support your answer.

```python
iris_setosa = iris[iris['species'] == 'setosa']
iris_versicolor = iris[iris['species'] == 'versicolor']
iris_virginica = iris[iris['species'] == 'virginica']

#calling each species and naming them as functions

setosa_corr = iris_setosa[['sepal_length','sepal_width']].corr()
versicolor_corr = iris_versicolor[['sepal_length','sepal_width']].corr()
virginica_corr = iris_virginica[['sepal_length','sepal_width']].corr()

#setting each species function to be a correlation between sepal length and sepal width, and naming each of those as a function

print("Setosa Correlation:", setosa_corr)
print("Versicolor Correlation:", versicolor_corr)
print("Virginica Correlation:", virginica_corr)

#printing each correlation function and labeling each

#based on the correlation values, the species with the highest correlation between sepal length and sepal width is the species setosa with a value of 0.742547
```

13.  Among the four variables, the two that are most negatively correlated (overall) are `sepal_width` and `petal_length`.  (How do we know this?)  Determine the same thing but for each species separately:  for each species, which two variables have the most negative (or the least positive) correlation?  Again, include calculations to support your answer.

```python
iris[['sepal_length','sepal_width', 'petal_length', 'petal_width']].corr()

#we know that sepal width and petal length are the most negatively correlated based on the iris correlation table here (those two have a correlation value of -0.428440)
```

```python
iris_setosa = iris[iris['species'] == 'setosa']
iris_versicolor = iris[iris['species'] == 'versicolor']
iris_virginica = iris[iris['species'] == 'virginica']

#again calling each species and naming them as functions

setosa_corr2 = iris_setosa[['sepal_length','sepal_width','petal_length', 'petal_width']].corr()
versicolor_corr2 = iris_versicolor[['sepal_length','sepal_width', 'petal_length', 'petal_width']].corr()
virginica_corr2 = iris_virginica[['sepal_length','sepal_width', 'petal_length', 'petal_width']].corr()

#setting each species function to be a correlation between all sepal and petal lengths and widths this time, and naming each of those as a function

print("Setosa Correlation:", setosa_corr2)
print("Versicolor Correlation:", versicolor_corr2)
print("Virginica Correlation:", virginica_corr2)

#For Setosa, the least positive correlation was between sepal_width and petal_length
#For Versicolor, the least positive correlation was between sepal_width and sepal_length
#For Virginica, the least positive correlation was between petal_width and sepal_length
```
