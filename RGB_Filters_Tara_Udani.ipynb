```python
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt
from skimage import io

url = 'https://s3.amazonaws.com/eit-planttoolbox-prod/media/images/Bellis_Perennis--Morgaine--CC_BY_2.0.jpg'
image = (Image.open(requests.get(url, stream=True).raw))

image
```

```python
def redfilter():    # apply function to get red filter
  global photo_red    # use 'global' to define photo_red outside the function
  photo = io.imread(url)   # photo variable to read url
  photo_red = photo.copy()
  photo_red[:,:,1] = 0  # sets green to 0 (R=0,G=1,B=2)
  photo_red[:,:,2] = 0  # sets blue to 0
  return photo_red  #for printing at the end

#repeat the process

def greenfilter():
  global photo_green
  photo = io.imread(url)
  photo_green = photo.copy()
  photo_green[:,:,0] = 0
  photo_green[:,:,2] = 0
  return photo_green

def bluefilter():
  global photo_blue
  photo = io.imread(url)
  photo_blue = photo.copy()
  photo_blue[:,:,0] = 0
  photo_blue[:,:,1] = 0
  return photo_blue

def redgreenfilter():
  global photo_redgreen
  photo = io.imread(url)
  photo_redgreen = photo.copy()
  photo_redgreen[:,:,2] = 0   #turns off only blue
  return photo_redgreen

def redbluefilter():
  global photo_redblue
  photo = io.imread(url)
  photo_redblue = photo.copy()
  photo_redblue[:,:,1] = 0   #turns off only green
  return photo_redblue

def greenbluefilter():
  global photo_greenblue
  photo = io.imread(url)
  photo_greenblue = photo.copy()
  photo_greenblue[:,:,0] = 0  #turns off only red
  return photo_greenblue

redfilter()
greenfilter()
bluefilter()
redgreenfilter()
redbluefilter()
greenbluefilter()

images =[photo_red, photo_green, photo_blue, photo_redgreen, photo_redblue, photo_greenblue]

# so the loop cycles
for ima in images:
  plt.figure()
  plt.imshow(ima)
```
